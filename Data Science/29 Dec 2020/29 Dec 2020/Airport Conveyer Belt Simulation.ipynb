```python
import csv
from csv import reader
from tabulate import tabulate
```

```python
class conveyer_belt:
    def read_csv(filename):
        with open(filename, 'r') as read_obj:
            csv_reader = reader(read_obj)
            list_of_passengers = list(csv_reader)
        columns = list_of_passengers.pop(0)
        return columns, list_of_passengers
        
    def belt_decider(data):
        c1 = []
        c2 = []
        c3 = []
        c4 = []
        for row in data:
            if int(row[1])<30 and int(row[1])>7 and (row[2]=='Small' or row[2]=='small' or row[2]=='Normal' or row[2]=='normal'):
                c1.append(row)
            elif int(row[1])==30 and (row[2]=='Small' or row[2]=='small' or row[2]=='Normal' or row[2]=='normal'):
                c2.append(row)
            elif int(row[1])>30 and (row[2]=='Normal' or row[2]=='normal' or row[2]=='Big' or row[2]=='big'):
                c3.append(row)
            elif int(row[1])<8 and row[2]=='Very Small':
                c4.append(row)
        return c1,c2,c3,c4
            
    def write_csv(column, list1, list2, list3, list4):
        list1 = [column]+list1
        list2 = [column]+list2
        list3 = [column]+list3
        list4 = [column]+list4
        file = open('C1.csv', 'w+', newline ='') 
        # writing the data into the file 
        with file:     
            write = csv.writer(file) 
            write.writerows(list1) 
        file.close()
        
        file = open('C2.csv', 'w+', newline ='') 
        # writing the data into the file 
        with file:     
            write = csv.writer(file) 
            write.writerows(list2)
        file.close()
        
        file = open('C3.csv', 'w+', newline ='') 
        # writing the data into the file 
        with file:     
            write = csv.writer(file) 
            write.writerows(list3)
        file.close()
        
        file = open('C4.csv', 'w+', newline ='') 
        # writing the data into the file 
        with file:     
            write = csv.writer(file) 
            write.writerows(list4)
        file.close()
       
```

```python
file_path = 'Assignment_data_input.csv'
col, passengers_data = conveyer_belt.read_csv(file_path)
belt1, belt2, belt3, belt4 = conveyer_belt.belt_decider(passengers_data)
conveyer_belt.write_csv(col, belt1, belt2, belt3, belt4)

val='0'
while int(val) != 5:
    print("MENU\nPlease select an option:")
    print("1.\tOpen conveyer belt 1\n2.\tOpen conveyer belt 2\n3.\tOpen conveyer belt 3\n4.\tOpen conveyer belt 4\n5.\tExit")
    val=input()
    if int(val)<5 and int(val)>0:
        filename = 'C'+val+'.csv'        
        col, conv_belt = conveyer_belt.read_csv(filename)

        print(tabulate(conv_belt, headers = col))
```
